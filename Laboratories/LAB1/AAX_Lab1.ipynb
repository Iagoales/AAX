```python
# Cell 3 – Plots/Grafics

import re, matplotlib.pyplot as plt, pandas as pd

def read_throughput(path):

    times, mbps = [], []
    with open(path) as f:
        for line in f:
            try:
                t = float(line.split(';', 1)[0])    #Extract time at the begging of the line
            except ValueError:
                continue

            m = re.search(r'Average throughput\s*=\s*([\d\.]+)', line)
            if m:
                times.append(t)
                mbps.append(float(m.group(1)))
    return times, mbps


plt.figure(figsize=(7,4))

for ap, path in AGENT_LOGS.items():        # AGENT_LOGS is defined in cell 2
    t, y = read_throughput(path)
    if not t:
        print(f"No data in{path}")
        continue
    pd.Series(y, index=t).sort_index().plot(label=ap)

plt.xlabel("Temps [s]")
plt.ylabel("Throughput [Mbps]")
plt.title("Temporal throughput per AP (monitor mode)")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Set up the data
egreedy_data = {
    'A': [38.520503],
    'B': [34.989296],
    'C': [37.833467]
}

softmax_data = {
    'A': [37.328643],
    'B': [23.757437],
    'C': [35.233216]
}

# Create DataFrames
egreedy_df = pd.DataFrame(egreedy_data)
softmax_df = pd.DataFrame(softmax_data)

# Plotting
plt.figure(figsize=(12, 6))

# Bar positions
bar_width = 0.35
index = np.arange(len(egreedy_df.columns))

# Create bars
bar1 = plt.bar(index - bar_width/2, egreedy_df.mean(), bar_width,
               label='ε-greedy', color='skyblue', edgecolor='black')
bar2 = plt.bar(index + bar_width/2, softmax_df.mean(), bar_width,
               label='Softmax', color='lightcoral', edgecolor='black')

# Customize the plot
plt.xlabel('Arms', fontsize=12)
plt.ylabel('Mean Reward', fontsize=12)
plt.title('Comparison of ε-greedy and Softmax Strategies', fontsize=14)
plt.xticks(index, egreedy_df.columns)
plt.legend()

# Add value labels on top of each bar
for bar in [bar1, bar2]:
    for rect in bar:
        height = rect.get_height()
        plt.text(rect.get_x() + rect.get_width()/2., height,
                f'{height:.2f}',
                ha='center', va='bottom')

plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
